#### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans -> The five plots that can be created using seaborn are as follows -

 1. Histogram -
 
 (sns.displot) It is used to visualise the freuency of a continious or discrete random variable.
 
 2. Lineplot  - 
 
 (sns.lineplot) It is used to create plots which show the trend of variable on y-axis vs variable on x-axis.
 
 3. Scatterplot -
 
 (sns.scatterplot) It is used to compare trends of variable on y vs x but each data point is represented by a dot.
 
 4. Boxplot   - 
 
 (sns.boxplot) It is used to visualize all the statistical parmeters like median, IQR , quartile and outliers.
 
 5. Barplot   - 
 
 (sns.barplot) Bar plots are a type of data visualization used to represent data in the form of rectangular bars.                 The height of each bar represents the value of a data point, and the width of each bar represents the category                   of the data. It helps to show the relationship between a numeric and a categoric variable.

#### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
#### parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
sns.displot(x = diamonds.price , bins = 8 , data = diamonds , hue = 'cut')
```

#### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import matplotlib.pyplot as plt
import numpy as np
df = sns.load_dataset('flights')
```

```python
df.head(12)
```

```python
matrix = []
i = 0
while(i<144):
        matrix.append(list(df['passengers'].iloc[i:i+12]))
        i = i+12
```

```python
matrix
```

```python
sns.heatmap(data = matrix,linewidth = 0.5,linecolor = 'black',xticklabels=df['month'].unique(),yticklabels=df['year'].unique())
plt.show()
```
