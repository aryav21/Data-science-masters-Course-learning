#### Q1. What is Flask Framework? What are the advantages of Flask Framework?

Ans -> Flask is a micro web framework written in Python that allows you to build web applications quickly and easily.Flask is built on top of the WSGI (Web Server Gateway Interface) toolkit and Jinja2 template engine, which provides a powerful and flexible environment for building web applications. Flask also supports a wide range of extensions, which makes it easy to add new functionality to your application.Flask is ideal for small to medium-sized projects, and it's particularly well-suited for building RESTful APIs and microservices.

Advantages:

- Lightweight: Flask is a lightweight framework that doesn't require particular tools or libraries, which means it's easy to learn and use.

- Easy to learn: Flask has a simple syntax and doesn't require much boilerplate code, making it easy to learn for developers new to web development.

- Flexibility: Flask is a flexible framework that allows developers to choose the components they want to use. This flexibility makes it easier to customize the application and add new functionality.

- Large community: Flask has a large and active community of developers who contribute to the framework and provide support to other developers.

- Compatibility: Flask is compatible with most Python libraries, making it easy to integrate other libraries into your application.

- Scalability: Flask allows for the creation of scalable applications that can handle large amounts of traffic.

#### Q3. What is App routing in Flask? Why do we use app routes?

Ans -> In Flask, routing refers to the process of mapping a URL to a view function.It allows you to define how different URLs should be handled by the application and which view function should be executed when a particular URL is requested.Flask provides a decorator called @app.route that allows you to define URL routes in your application. You can specify the URL pattern and the HTTP methods that should be allowed for each route.

Reasons for using app routes are as follows -

- Handling user requests: App routes allow you to map specific URLs to the appropriate view functions. This allows you to handle user requests and generate appropriate responses.

- Creating RESTful APIs: App routes are often used to create RESTful APIs in Flask. Each URL maps to a specific resource or endpoint, and the corresponding view function handles requests for that resource.

- Separating concerns: By using app routes, you can separate different parts of your application into different routes, each with its own view function. This makes it easier to maintain and modify your application over time.

- URL generation: App routes can be used to generate URLs dynamically in your templates. This allows you to create dynamic links to different parts of your application, making it easier for users to navigate.

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")   ## url for the homepage of application
def hello_world():
    return "<h1>Hello World!!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

#### url_for() function.

Ans -> The url_for() function is used to build a URL to the specific function dynamically. The first argument is the name of the specified function, and then we can pass any number of keyword argument corresponding to the variable part of the URL.

This function is useful in the sense that we can avoid hard-coding the URLs into the templates by dynamically building them using this function.Flask url_for is defined as a function that enables developers to build and generate URLs on a Flask application. As a best practice, it is the url_for function that is required to be used, as hard coding the URL in templates and view function of the Flask application tends to utilize more time during modification. If we are using hard coding, and in case we would need to change our URL by inserting another element in the URL, we would have to visit each and every template or form in the code library and make the modifications and will lead to overkill. The url_for function is capable of changing this with just a snap of fingers
